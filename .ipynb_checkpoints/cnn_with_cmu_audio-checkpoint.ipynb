```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.cnn_layers = torch.nn.Sequential(
            torch.nn.Conv2d(1, 8, kernel_size=3),
            torch.nn.ReLU(),
            torch.nn.Conv2d(8, 16, kernel_size=3),
            torch.nn.ReLU()
        )
        
        self.linear_layers = torch.nn.Linear(16 * 24 * 24, 10)
        
    def forward(self, x):
        x = self.cnn_layers(x)
        x = torch.flatten(x, 1)
        x = self.linear_layers(x)
        return x
```
